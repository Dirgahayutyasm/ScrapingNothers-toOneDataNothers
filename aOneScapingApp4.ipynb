## Url untuk scraping nama mahasiswa bimbingan, setelah dosen berhasil login

```python
urlScrap = ['pra','p0','p1','p2','semhas']

urlScrap1 = "https://filkom.ub.ac.id/module/skripsi/detail/pembimbing/pra"
#driver.get(urlScrap1)

##python_button = driver.find_element_by_id('MainContent_uxLevel2_JobTitles_uxJobTitleBtn_' + str(x))
#WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,"//a[@class='nav-link']"))).click()
##python_button.click()

# navigate other page yg baru dapat diakses setelah login
driver.get(urlScrap1)

#Sleep untuk menunggu loading data (delay 10 detik)
SET_LOAD_TIME = 10
time.sleep(SET_LOAD_TIME)
print("Setelah delay 10 detik")

soup_level2=BeautifulSoup(driver.page_source, 'lxml')

print()
print('Title: "{}"'.format(driver.title))
print()
print(soup_level2)

print()
init_get_jml_bim = soup_level2.find_all("div", attrs={"id": "data-pra_info"})
print(init_get_jml_bim)


```

```python
print(type(init_get_jml_bim[0]))
print(init_get_jml_bim[0].text)
get_jml_bim_tampil = init_get_jml_bim[0].text.split(' ')
print(get_jml_bim_tampil[3])
print(get_jml_bim_tampil[5])

get_jml_bim_page1 = get_jml_bim_tampil[3]
get_jml_bim_total = get_jml_bim_tampil[5]

# menentukan jumlah klik next
if int(get_jml_bim_page1) <= int(get_jml_bim_total):
    jml_klik_next = 0
else:
    jml_klik_next = int(-(-float(get_jml_bim_total)//float(10.)))
print(jml_klik_next)
print()

#get nama mhs bimbingan dari Web Filkom
#get_nama = soup_level2.find_all("span", attrs={"class": "tx-inverse"})[4:-2]
get_nama = []
for item in soup_level2.find_all('span',class_="tx-inverse"): 
    if len(item["class"]) != 1: 
        continue;
    else:
        #get_nama.append(item.text)
        get_nama.append((' '.join(sub[:1].upper() + sub[1:] for sub in item.text.lower().split(' '))).lstrip())
print(get_nama)

```
