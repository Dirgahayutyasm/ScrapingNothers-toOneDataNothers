```python
def getMaxAkurasi(k,fold_ke):
    lines_modif = []
    namefilelog='datalog/autoRunKNN-' + str(k) + '-' + str(fold_ke) + '.log'
    print(namefilelog)
    for line in open(namefilelog):
        if "Akurasi Terbesar" in line:
            lines_modif.append(line)
    #print(lines_modif[0])
    print(len(lines_modif))
    print()

    import matplotlib.pyplot as plt
    test_i_th_modif = []
    accuracy_save = []

    print('Retrieving data and plotting testing accuracy graph k-Fold ' + str(k))
    for i in range(len(lines_modif)):
        lineParts = lines_modif[i].split(',')
        test_i_th_modif.append(i+1)
        accuracy_save.append(float(lineParts[0].split()[3]))

    # write avg_loss_modif
    accfilelog='accuracy-autoRunKNN-' + str(k) + '-' + str(fold_ke) + '.txt'
    with open(accfilelog, 'w') as f:
        for item in accuracy_save:
            f.write("%s\n" % item)
    f.close()
    
    iterfilelog='test_i_th-autoRunKNN-' + str(k) + '-' + str(fold_ke) + '.txt'
    with open(iterfilelog, 'w') as f:
        for item in test_i_th_modif:
            f.write("%s\n" % item)

    f.close()
    
    #print(test_i_th_modif)
    #print(test_i_th_modif[0:2])
    
    #test_i_th_modif=list(range(10,304))

    fig = plt.figure()
    for i in range(0, len(lines_modif)):
        plt.plot(test_i_th_modif[i:i+2], accuracy_save[i:i+2], 'b.-')
    #plt.plot(test_i_th_modif, accuracy_save, 'r.-')

    plt.style.use('ggplot')
    plt.xlabel('Test i-th (k-Fold = ' + str(k) +')')
    plt.ylabel('Accuracy (%)')
    plt.show()

    print(max(accuracy_save))
    return max(accuracy_save)
```

```python
accuracy_max_save = []
for i in range(3,11):    
    accuracy_max_save.append(getMaxAkurasi(i,10))
```

```python
print(accuracy_max_save)
```

```python
k_fold=list(range(3,11))
fig = plt.figure()
for i in range(0, len(accuracy_max_save)):
    plt.plot(k_fold[i:i+2], accuracy_max_save[i:i+2], 'g.-')

plt.xlabel('k-Fold')
plt.ylabel('Accuracy (%)')
plt.show()
```
